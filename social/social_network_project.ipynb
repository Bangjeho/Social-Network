```python
USA_data.to_csv('USA_Company.csv',header = True, index = False)
```

```python
USA_data['Category'] = 'Twitter'
```

```python
for i in range(0,len(USA_data)):
    if (USA_data['ParentCompany'][i] == 'Twitter' or
        USA_data['ParentCompany'][i] == 'Facebook'):
        
        USA_data['Category'][i] = "SNS"
        
    elif (USA_data['ParentCompany'][i] == 'Google' or 
        USA_data['ParentCompany'][i] == 'Yahoo'):
        USA_data['Category'][i] = "Internet_Search"
        
    else :
        USA_data['Category'][i] = "Computing_platform"
```

```python
USA_data.to_csv('USA_Company.csv',header = True, index = False)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    
    plt.hist(degrees)
    plt.show()
    
er = nx.erdos_renyi_graph(30,0.5) #n,p
ws = nx.watts_strogatz_graph(30,2,0.5) #n,k,p
ba = nx.barabasi_albert_graph(30,2) #n,m
```

```python
plot_degree_dist(er)
```

```python
plot_degree_dist(ws)
```

```python
plot_degree_dist(ba)
```

```python
nx.write_pajek(er, "er.net")
nx.write_pajek(ws, "ws.net")
nx.write_pajek(ba, "ba.net")
```

```python
options = {'node_color' : "black",
          'node_size'  :5,
          'width' : 1
          }
plt.subplot(231)
nx.draw_kamada_kawai(er,**options)
plt.subplot(232)
nx.draw_kamada_kawai(ws,**options)
plt.subplot(233)
nx.draw_kamada_kawai(ba,**options)
plt.subplot(234)
nx.draw_circular(er,**options)
plt.subplot(235)
nx.draw_circular(ws,**options)
plt.subplot(236)
nx.draw_circular(ba,**options)
```
